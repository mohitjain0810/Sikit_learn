```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("500hits.csv", encoding = 'latin-1')
df.head()
```

```python
X = df.drop(columns = ['PLAYER', 'HOF'])
```

```python
y = df['HOF']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 11, test_size = 0.2)
```

This line splits your data into training and testing sets.

X â†’ Features (independent variables)

y â†’ Target (dependent variable)

âœ… train_test_split() splits the data:

X_train â†’ Training data (features)

X_test â†’ Testing data (features)

y_train â†’ Training labels (target)

y_test â†’ Testing labels (target)

test_size = 0.2 means:

âœ… 20% of the data will be used for testing.
âœ… 80% of the data will be used for training.

ðŸ’¡ Example:

If you have 1000 rows,

800 rows â†’ Training set

200 rows â†’ Testing set

ðŸ“Œ random_state=11 â†’ Ensures consistent results every time you run the code.
