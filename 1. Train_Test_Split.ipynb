```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("500hits.csv", encoding = 'latin-1')
df.head()
```

```python
X = df.drop(columns = ['PLAYER', 'HOF'])
```

```python
y = df['HOF']
```

```python
X.head()
```

```python
y.head()
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 11, test_size = 0.2)
```

This line splits your data into training and testing sets.

X → Features (independent variables)

y → Target (dependent variable)

✅ train_test_split() splits the data:

X_train → Training data (features)

X_test → Testing data (features)

y_train → Training labels (target)

y_test → Testing labels (target)

test_size = 0.2 means:

✅ 20% of the data will be used for testing.
✅ 80% of the data will be used for training.

💡 Example:

If you have 1000 rows,

800 rows → Training set

200 rows → Testing set

📌 random_state=11 → Ensures consistent results every time you run the code.
